```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

a=input("Write the ticker of Company 1: ")
b=input("Write the ticker of Company 2: ")
c=input("Name of company 1: ")
d=input("name of company 2: ")

stock_symbol1=a
stock_symbol2=b

# Download historical data
stock1 = yf.Ticker(stock_symbol1)
stock2 = yf.Ticker(stock_symbol2)

data1 = stock1.history(period="1y")
data2 = stock2.history(period="1y")



# Extract relevant information
eps1 = stock1.info['trailingEps']
eps2 = stock2.info['trailingEps']


pe1 = stock1.info['trailingPE']
pe2 = stock2.info['trailingPE']

mcap1 = stock1.info['marketCap']
mcap2 = stock2.info['marketCap']

# Plot the closing prices for both stocks
plt.figure(figsize=(12, 6))
plt.plot(data1.index, data1['Close'], label=stock_symbol1)
plt.plot(data2.index, data2['Close'], label=stock_symbol2)
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price Comparison')
plt.legend()
plt.grid(True)
plt.show()


#COMPARISON

if eps1>eps2:
  print(c,"has higher EPS")
else:
  print(d,"has higher EPS")

if pe1>pe2:
  print(c,"has higher Profit Earning ratio")
else:
  print(d,"has higher Profit Earning ratio")

if mcap1>mcap2:
  print(c,"has higher market capitalization")
else:
  print(d,"has higher market capitalization")

if eps1>eps2 and pe1>pe2:
  print(d,"is highly profitable")
elif pe1>pe2 and mcap1>mcap2:
  print(d,"is highly profitable")
elif eps1>eps2 and mcap1>mcap2:
  print(d,"is highly profitable")

if eps2>eps1 and pe2>pe1:
  print(c,"is highly profitable")
elif pe2>pe1 and mcap2>mcap1:
  print(c,"is highly profitable")
elif eps2>eps1 and mcap2>mcap1:
  print(c,"is highly profitable")

#To check the volatility of Company 1

historical_data = stock1.history(period="1y")
v1 = np.std(historical_data["Open"])
v2=np.std(historical_data["Close"])
v3=np.std(historical_data["High"])
v4=np.std(historical_data["Low"])
arr=[v1,v2,v3,v4]
va=np.mean(arr)

#To check the volatility of Company 2

historical_data = stock2.history(period="1y")
v1 = np.std(historical_data["Open"])
v2=np.std(historical_data["Close"])
v3=np.std(historical_data["High"])
v4=np.std(historical_data["Low"])
arr=[v1,v2,v3,v4]
vb=np.mean(arr)

if va>vb:
  print("Volatility of",c,"is higher than",d)
else:
  print("Volatility of",d,"is higher than",c)







```
